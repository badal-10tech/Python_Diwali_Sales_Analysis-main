# Exploratory Data Analysis

# Gender

```python
sns.despine() # to remove the border (spines) around a plot
```

```python
sns.set_theme(style='ticks')
ax = sns.countplot(data = df,x = 'Gender',hue = 'Gender' ,palette='dark:b')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby('Gender',as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
sales_gen
```

```python

ax = sns.barplot(data = sales_gen , x = 'Gender' , y = 'Amount' ,hue = 'Gender' )
for bars in ax.containers:
    ax.bar_label(bars ,fmt='%.0f')
plt.title("Female sales amount vs Male sales amount")
plt.show()
```

**From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men**

# Age

```python
ax = sns.countplot(data = df , x = "Age Group" , hue = 'Gender' )
for bars in ax.containers:
    ax.bar_label(bars)
    
```

**Total amount vs Age group**

```python
df_plot = df.groupby('Age Group',as_index = False)['Amount'].sum().sort_values( by = 'Amount',  ascending = False)
```

```python
ax = sns.barplot(data = df_plot, x= 'Age Group' , y = 'Amount' )
for bars in ax.containers:
    ax.bar_label(bars , fmt= '%.0f')
plt.show()
```

**From above graphs we can see that most of the buyers are of age group between 26-35 yrs female**

# States

```python
df.info()
```

```python
# Total number of orders from top 10 states

df_state_Sales_plot = df.groupby('State',as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending =False).head(10)
df_state_Sales_plot
```

```python
sns.set(rc={'figure.figsize' : (15,5)})
ax = sns.barplot(data =df_state_Sales_plot,x ='State', y ="Orders" ,hue = 'State',palette='dark:b')
for bars in ax.containers:
    ax.bar_label(bars ,fmt= '%.0f' )
```

```python
#  Total amount/sales of top 10 states
df_state_Sales_amount = df.groupby('State',as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending =False).head(10)
df_state_Sales_amount
```

```python
sns.set(rc={'figure.figsize' : (15,5)})
ax = sns.barplot(data =df_state_Sales_amount,x ='State', y ="Amount" ,hue = 'State',palette='dark:b')
for bars in ax.containers:
    ax.bar_label(bars ,fmt= '%.0f' )
```

**From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively**

# Product Category 

```python
sns.set(rc = {'figure.figsize' : (20,5)})
ax = sns.countplot(data= df ,x= 'Product_Category',order=df['Product_Category'].value_counts().index,hue='Product_Category',palette='dark:b')
for bars in ax.containers:
    ax.bar_label(bars ,fmt='%.0f')
```

```python
df_plot = df.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False ).head(10)
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(data= df_plot , x ='Product_Category',y = 'Amount',hue='Product_Category',palette="dark:b" )
ax.get_ylim()
for bars in ax.containers:
    ax.bar_label(bars,fmt = "%.0f")

```

***From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category***

```python
total_orders = df.groupby('Product_Category',as_index=False)['Orders'].sum().sort_values(by ='Orders',ascending=False ).head(10)
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(data= total_orders , x ='Product_Category',y = 'Orders',hue='Product_Category',palette="dark:b" )
plt.ylabel("orders_count")
ax.get_ylim()
for bars in ax.containers:
    ax.bar_label(bars,fmt = "%.0f")
```

# Conclution

Married women age group 26-35yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food Clothing and Electronics category
